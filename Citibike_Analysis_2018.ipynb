```python
# Dependencies
import pandas as pd
import glob
import datetime
```

```python
# Merging citibike_trip csv files (without repeating header)
citibike_trip_files = glob.glob("citibike_trip/*.csv") 
citibike_trip_df = pd.concat((pd.read_csv(f, header = 0) for f in citibike_trip_files))
```

```python
citibike_trip_df
```

```python
# Add new column extracting month_year for further analysis 
citibike_trip_df['month_year'] = pd.to_datetime(citibike_trip_df['starttime']).dt.to_period('M')
citibike_trip_df.head()
```

```python
# Reset Index
citibike_trip_df = citibike_trip_df.reset_index(drop=True)
citibike_trip_df.head(10)
```

```python
# Export citibike_trip_df file as a CSV
citibike_trip_df.to_csv("2018_citibike_trip_history_report.csv")
```

```python
citibike_trip_history_df = pd.read_csv("citibike_trip/2018_citibike_trip_history_report.csv")
trips_df = citibike_trip_history_df.groupby(['start station name','end station name']).size().reset_index(name = 'number of trips')
```

```python
trips_df
```

```python
# Export trips_df file as a CSV
trips_df.to_csv("citibike_trip/2018_citibike_popular_trips_report.csv")
```
